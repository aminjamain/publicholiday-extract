```python
import pandas as pd
from bs4 import BeautifulSoup
from tabulate import tabulate  # For better visual display of data

print("Reading data from file...")
pdver = pd.__version__
print("Pandas version: ", pdver)
import bs4
bs4ver = bs4.__version__
print("BeautifulSoup version: ", bs4ver)
```

Read from Path

```python
# Path to the local HTML file
html_file_path = "/Users/aminjamain/Documents/Phyton-Project/publicholiday-extract/Malaysia Public Holidays 2025.html"

# Read the HTML file
with open(html_file_path, "r", encoding="utf-8") as file:
    soup = BeautifulSoup(file, "html.parser")

# Extract the table
table = soup.find("table")
print("Table extracted successfully!")
```

Extract and print all the the data for inspection

```python
holiday_df = pd.read_html(str(table))[0]
print("Dataframe created successfully!")
holiday_df.columns = ["Date",
                      "Day",
                      "Holiday",
                      "State"]
print("Columns renamed successfully!")
table = tabulate(holiday_df, headers='keys', tablefmt='psql')
print(table)
```
